```python
'''Date: 7/11/2025

Lab - 4

Name: Rontala Pooja Reddy

HTNO: 2503B05130

Mtech 1st year

Topic: Tower of Hanoi'''

def tower_of_hanoi(n, source, target, auxiliary):
    if n == 1:
        print(f"Move disk 1 from {source} to {target}")
        return 1
    steps = tower_of_hanoi(n - 1, source, auxiliary, target)
    print(f"Move disk {n} from {source} to {target}")
    steps += 1
    steps += tower_of_hanoi(n - 1, auxiliary, target, source)
    return steps

n = int(input("Enter the number of disks: "))
print(f"\nSteps to solve Tower of Hanoi with {n} disks:\n")
total_steps = tower_of_hanoi(n, 'A', 'C', 'B')
print(f"\nTotal number of steps: {total_steps}")
```
