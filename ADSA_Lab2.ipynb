```python
class PatientNode:
    def __init__(self, name, age, pid):
        self.name = name
        self.age = age
        self.pid = pid
        self.next = None


class CircularPatientList:
    def __init__(self):
        self.last = None

    def insert_patient(self, name, age, pid):
        new_node = PatientNode(name, age, pid)

        if self.last is None:
            self.last = new_node
            self.last.next = self.last
        else:
            new_node.next = self.last.next
            self.last.next = new_node
            self.last = new_node

        print(f"\nPatient {name} (ID: {pid}) added successfully.")

    def delete_patient(self, pid):
        if self.last is None:
            print("\nNo patients to delete.")
            return

        current = self.last.next
        prev = self.last

        while True:
            if current.pid == pid:
                if current == self.last and current.next == self.last:
                    self.last = None
                elif current == self.last:
                    prev.next = current.next
                    self.last = prev
                else:
                    prev.next = current.next

                print(f"\nPatient with ID {pid} deleted successfully.")
                return

            prev = current
            current = current.next

            if current == self.last.next:
                print(f"\nPatient with ID {pid} not found.")
                break

    def display_patients(self):
        if self.last is None:
            print("\nNo patients in the list.")
            return

        current = self.last.next
        print("\nPatient List (Round-Robin Order):")
        while True:
            print(f"Name: {current.name}, Age: {current.age}, ID: {current.pid}")
            current = current.next
            if current == self.last.next:
                break

    def round_robin_checkup(self, rounds):
        if self.last is None:
            print("\nNo patients available for check-up.")
            return

        current = self.last.next
        print("\nRound Robin Check-Up:")
        for i in range(1, rounds + 1):
            print(f"Round {i}: Checking up Patient {current.name} (ID: {current.pid})")
            current = current.next


if __name__ == "__main__":
    plist = CircularPatientList()

    while True:
        print("\nPatient Round-Robin System")
        print("1. Insert Patient")
        print("2. Delete Patient")
        print("3. Display Patients")
        print("4. Round-Robin Check-Up")
        print("5. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            name = input("Enter name: ")
            age = int(input("Enter age: "))
            pid = int(input("Enter ID: "))
            plist.insert_patient(name, age, pid)

        elif choice == "2":
            pid = int(input("Enter patient ID to delete: "))
            plist.delete_patient(pid)

        elif choice == "3":
            plist.display_patients()

        elif choice == "4":
            rounds = int(input("Enter number of rounds: "))
            plist.round_robin_checkup(rounds)

        elif choice == "5":
            print("\nExiting... Thank you!")
            break

        else:
            print("\nInvalid choice! Try again.")
```
