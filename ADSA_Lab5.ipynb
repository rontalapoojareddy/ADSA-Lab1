```python
def fib(n):
  if n==0 or n==1:
    return n
  return fib(n-1)+fib(n-2)
print(fib)
```

```python
#recursion using stacks
def  fib_stack(n):
  stack=[]
  while stack:
    current_n,state=stack.pop()
    if state==0:
      if current_n==0 or current_n==1:
        return current_n
      stack.append((current_n,state+1))
      stack.append((current_n-1,0))
```
