```python
def fib(n):
  if n==0 or n==1:
    return n
  return fib(n-1)+fib(n-2)
print(fib(10) )
```

```python
def recursive_stack_processor(stack):
    """Recursively processes items by popping them from a list, simulating a stack."""
    if not stack:
        print("Stack is empty, recursion ends.")
        return

    item = stack.pop()  # Pop the last element (LIFO for list as stack)
    print(f"Processing item: {item}")

    # Recursive call with the modified stack
    recursive_stack_processor(stack)

# Example Usage:
my_stack = [10, 20, 30, 40, 50]
print("Original stack:", my_stack)

recursive_stack_processor(my_stack)
print("Stack after processing:", my_stack)
```

```python
def fib_stack(n):
    if n < 0:
        raise ValueError("Input must be a non-negative integer")
    if n <= 1:
        return n

    # Use a dictionary for memoization to store computed Fibonacci values
    memo = {0: 0, 1: 1}

    # The stack will store numbers whose Fibonacci value we need to compute
    computation_stack = [n]

    while computation_stack:
        current_n = computation_stack[-1]  # Peek at the top of the stack

        if current_n in memo:
            computation_stack.pop()  # Value already computed, remove from stack
            continue

        # If not in memo, we need to compute it. This requires fib(current_n - 1) and fib(current_n - 2).
        child1 = current_n - 1
        child2 = current_n - 2

        # Check if both children's Fibonacci values are already known
        if child1 in memo and child2 in memo:
            memo[current_n] = memo[child1] + memo[child2]
            computation_stack.pop()  # Current_n can now be computed and removed
        else:
            # If not both children are known, push them onto the stack (if not already there/memoized)
            # We push the larger child first to ensure smaller values are at the top for processing order
            if child2 not in memo and child2 >= 0:
                computation_stack.append(child2)
            if child1 not in memo and child1 >= 0:
                computation_stack.append(child1)
            # The current_n stays on the stack until its children are resolved

    return memo[n]

# Example Usage:
print(fib_stack(0)) # 0
print(fib_stack(1)) # 1
print(fib_stack(2)) # 1
print(fib_stack(3)) # 2
print(fib_stack(10)) # 55
```
