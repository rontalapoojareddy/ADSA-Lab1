```python
#2503B05130
# M.Tech_CSE
# R.Pooja Reddy

class CircularQueue:
    def __init__(self, size):
        self.size = size
        self.queue = [None] * size
        self.front = -1
        self.rear = -1
        self.job_count = 0

    def is_full(self):
        return self.job_count == self.size

    def is_empty(self):
        return self.job_count == 0

    def enqueue(self, job_name):
        if self.is_full():
            print(f"Queue Full. Cannot enqueue job '{job_name}'.")
            return

        if self.is_empty():
            self.front = 0

        self.rear = (self.rear + 1) % self.size
        self.queue[self.rear] = job_name
        self.job_count += 1
        print(f"Enqueued Job: {job_name}")

    def dequeue(self):
        if self.is_empty():
            print("Queue Empty. Cannot dequeue job.")
            return

        job = self.queue[self.front]
        self.queue[self.front] = None

        if self.job_count == 1:
            self.front = -1
            self.rear = -1
        else:
            self.front = (self.front + 1) % self.size

        self.job_count -= 1
        print(f"Dequeued Job: {job}")

    def display(self):
        if self.is_empty():
            print("Queue is Empty.")
            return

        print("\n--- Printer Jobs in Queue ---")
        index = self.front
        for _ in range(self.job_count):
            print(self.queue[index])
            index = (index + 1) % self.size
        print("-------------------------------")


# âœ… MAIN PROGRAM (this part was missing)
size = int(input("Enter Queue Size: "))
cq = CircularQueue(size)

while True:
    print("\n1. Enqueue Job")
    print("2. Dequeue Job")
    print("3. Display Queue")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        job = input("Enter job name: ")
        cq.enqueue(job)

    elif choice == 2:
        cq.dequeue()

    elif choice == 3:
        cq.display()

    elif choice == 4:
        print("Exiting program...")
        break

    else:
        print("Invalid choice! Try again.")
```
